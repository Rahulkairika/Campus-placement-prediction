```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
data = data.drop('StudentID', axis=1)
```

```python
data.head()
```

```python
data['ExtracurricularActivities']=le.fit_transform(data['ExtracurricularActivities'])
data['PlacementTraining']=le.fit_transform(data['PlacementTraining'])
data['PlacementStatus']=le.fit_transform(data['PlacementStatus'])
```

```python
data.head(10)
```

```python
sns.boxplot(data['PlacementStatus'])
```

```python
y=data['PlacementStatus']
X=data.drop(columns=['PlacementStatus'])
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale=StandardScaler()
```

```python
names=X.columns
```

```python
X=scale.fit_transform(X)
```

```python
X
```

```python
X=pd.DataFrame(X,columns=names)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X ,y ,test_size=0.3,random_state=42)
```

```python
train_test_split
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#classification
#Logisticregression
from sklearn.linear_model import LogisticRegression
```

```python
logi=LogisticRegression()
```

```python
logi.fit(x_train,y_train)
```

```python
ytr=logi.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
print(accuracy_score(y_train,ytr))
```

```python
pred=logi.predict(x_test)
cm=confusion_matrix(y_test,pred)
sns.heatmap(cm,annot=True)
```
